```python
import pandas as pd
import numpy as np

fake_df = pd.read_csv("../data/misinfo/fake.csv")
real_df = pd.read_csv("../data/misinfo/real.csv")

print(fake_df.shape)
print(real_df.shape)

fake_df["label"] = 0   # Fake / Misleading
real_df["label"] = 1   # Real

fake_df.columns
real_df.columns
fake_df = fake_df[["text", "label"]]
real_df = real_df[["text", "label"]]

df = pd.concat([fake_df, real_df], ignore_index=True)

df.head()
df.tail()
df.shape

df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

df.reset_index(drop=True, inplace=True)
df.isnull().sum()

df["label"].value_counts()

df.to_csv("../data/clean_misinfo.csv", index=False)
```
